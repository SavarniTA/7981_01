```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("diamonds.csv")
```

```python
len(df[df.duplicated()])
```

```python
df.dropna(subset=['carat','cut'])
```

```python
df._get_numeric_data()
```

```python
df['z'].fillna(0, inplace = True) 
  
```

```python
for i in range(0,len(df)):
    if(df['z'][i]=="None"):
        df['z'][i]=0
```

```python
for i in range(0,len(df)):
    df['z']=df['z'].astype(float)
   
```

```python
ans=[]
for i in range(0,len(df)):
    if df['depth'][i]>60:
        ans.append(df['x'][i]*df['y'][i]*df['z'][i])
    else:
        ans.append(8)
df['volume']=ans
```
