```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("imdb.csv",escapechar="\\")
df.to_csv("imdb2.csv",index=False)
```

```python
def fifth_movie(df):
    h=df['imdbRating'][4]
    return h
```

```python
fifth_movie(df)
```

```python
def movies(df):
    minval=df[df["duration"]==df["duration"].min()]["title"]
    maxval=df[df["duration"]==df["duration"].max()]["title"]
    return minval,maxval
```

```python
movies(df)
```

```python
def sort_df(df):
    so=df.sort_values(by=['year', 'imdbRating'],ascending = [True,False])
    return so
```

```python
sort_df(df)
```

```python
df=pd.read_csv("movie_metadata.csv")
```

```python
def subset_df(df):
    sa=df[(df['gross'].astype(float)>2000000)&(df['budget'].astype(float)<1000000) & (df['duration'].astype(float)>30) &    (df['duration'].astype(float)<180)]
    return sa
```
