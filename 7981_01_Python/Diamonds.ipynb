```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("diamonds.csv")
```

```python
def dupl_rows(df):
    q=len(df[df.duplicated()])
    return q
```

```python
dupl_rows(df)
```

```python
def drop_row(df):
    hs=df.dropna(subset=['carat','cut'])
    return hs
```

```python
drop_row(df)
```

```python
def sub_numeric(df):
    num=df._get_numeric_data()
    return num
```

```python
sub_numeric(df)
```

```python
def volume(df):
    df['z'].fillna(0, inplace = True)
    for i in range(0,len(df)):
        if(df['z'][i]=="None"):
            df['z'][i]=0
    for i in range(0,len(df)):
        df['z']=df['z'].astype(float)
    ans=[]
    for i in range(0,len(df)):
        if df['depth'][i]>60:
            ans.append(df['x'][i]*df['y'][i]*df['z'][i])
        else:
            ans.append(8)
    df['volume']=ans
    return df
```

```python
volume(df)
```

```python
def impute(df):
    df=df.fillna(df.mean())
    return df
```

```python
impute(df)
```
