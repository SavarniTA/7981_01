```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("imdb.csv",escapechar="\\")
```

```python
df['GenreCombo']=df[df.columns[16:]].T.apply(lambda g: '|'.join(g.index[g==1]),axis=0)
df.groupby(["type","year","GenreCombo"]).agg({"imdbRating":[min,max,np.mean],'duration':np.sum})
```

```python
df['Title_Length']=df['title'].str.len()
```

```python
df['Quantile']=pd.qcut(df['Title_Length'],4,labels=False)
dataf=pd.crosstab(df.year,df.Quantile,margins=False)
dataf["min"]=df.groupby(["year"])["Title_Length"].min()
dataf["max"]=df.groupby(["year"])["Title_Length"].max()
```

```python
df=pd.read_csv("diamonds.csv")
```

```python
df['z'].fillna(0, inplace = True) 
for i in range(0,len(df)):
    if(df['z'][i]=="None"):
        df['z'][i]=0
```

```python
for i in range(0,len(df)):
    df['z']=df['z'].astype(float)
```

```python
ans=[]
for i in range(0,len(df)):
    if df['depth'][i]>60:
        ans.append(df['x'][i]*df['y'][i]*df['z'][i])
    else:
        ans.append(8)
df['volume']=ans
```

```python
df["bins"]=pd.qcut(df["volume"],q=5,labels=['1','2','3','4','5'])
```

```python
df3=(pd.crosstab(df["bins"],df["cut"],normalize=True))*100
```

```python
df3
```

```python
df['decile'] = pd.qcut(df['duration'], 10, labels=False)
df.groupby(["decile"]).agg({"nrOfNominations":np.sum,"nrOfWins":np.sum,"ratingCount":np.sum})
```

```python
datf=pd.crosstab(df.decile,df.GenreCombo,margins=False)
```
